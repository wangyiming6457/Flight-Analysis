```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d1 = pd.read_csv("../input/new-dataset/2007.csv")
d2 = pd.read_csv("../input/new-dataset/2008.csv")
```

```python
d1.head()
```

```python
d2.head()
```

```python
combined = [d1,d2]
dataset = pd.concat(combined)
```

```python
len(dataset.columns)
```

```python
dataset.head()
```

```python
byDayOfMonth = dataset.groupby(['DayOfWeek'], as_index=False).agg({'DayOfWeek': 'count'})
byDayOfWeek = dataset.groupby(['DayOfWeek'], as_index=False).agg({'DayOfWeek': 'count'})
byMonth = dataset.groupby(['Month'], as_index=False).agg({'Month': 'count'})
```

```python
monthNames = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
byMonth["Names"] = monthNames
```

```python

byMonth = byMonth.rename({'Month': 'Flights'}, axis=1)
```

```python
byMonth.head()
```

```python
sns.catplot(x='Names', y='Flights',data=byMonth, kind='bar')
plt.title("Monthly Average Departure Delay") 
plt.plot()
plt.show()
```

From the above plot, that has flights for each month. We see that more flights are book in the month of Jan, Feb, Mar and Apr in year of 2007-08. But, we see almost equal number of flights in the rest of months for year 2007-08. Hence, we can say that the beginner 4 fours moths where more people travel.

```python
flights_By_Origin = dataset.groupby(['Month', 'Origin']).size().reset_index(name='counts')
```
