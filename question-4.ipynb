```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d1 = pd.read_csv("../input/new-dataset/2007.csv")
d2 = pd.read_csv("../input/new-dataset/2008.csv")
```

```python
combined = [d1,d2]
dataset = pd.concat(combined)
```

```python
dataset.columns
```

```python
important_features= ["DayofMonth","TailNum","Origin","Dest","CRSDepTime","DepTime","DepDelay"]
indices = [i for i,feature in enumerate(dataset.columns) if feature in important_features]
```

```python
newdataset = dataset.iloc[:,indices]
newdataset.sort_values(by='DayofMonth', ascending=True)
```
