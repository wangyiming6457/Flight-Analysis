```python
# When is the best time of day, day of the week, and time of year to fly to minimise delays?
```

To find the best time of the data, day of the week and time of the year so that we can minimise delays. Let's first fetch the necessary columns.

```python
new_features = ["Year",'Month', 'DayofMonth',"DayOfWeek","DepTime"]
delay_features = [value for value in features if "Delay" in value]
new_features = new_features + delay_features
```

```python
newdataset = dataset[new_features]
```

Now, we have extracted the necessary columns including all the delay columns.

```python
newdataset.head()
```

```python
delay_features
```

Now, let take the mean of all the delay columns by rows.

```python
newdataset["AverageDelay"] = newdataset[delay_features].mean(axis=1)
```

```python
newdataset.head()
```

# Year with Minimum Delay

1. We have calculated the average delay by year to find year with minimum delay.

```python
bestYear = pd.DataFrame(newdataset.groupby(["Year"], as_index=False).mean().groupby('Year')['AverageDelay'].mean())
```

```python
bestYear
```

We can clearly see that year 2008 has mean delay of 3.56 mins as compared to 2007 which hs around average delay of 5 mins.

# Day Of Week with Minimum Delay

```python
bestDay = pd.DataFrame(newdataset.groupby(["Year","DayOfWeek"], as_index=False).mean().groupby('DayOfWeek')['AverageDelay'].mean())
```

```python
bestDay
```

```python
sns.scatterplot(data=bestDay,x="DayOfWeek",y="AverageDelay")
```

We can see that 6th day of the week has lowest average delay from year 2007-2008. Hence, we can say that day 6 has the minimum delay.

# Best Dep Time of Year

```python
sns.scatterplot(data=newdataset,x="DepTime",y="AverageDelay",hue="Year",palette="deep")
```

Above scatterplot defines Average Delay by Departure Time. From 0th hour, the delay decreases gradually till 500 hour and then starts increasing gradualy with few spikes around 1200. We can say that we have quite lower delays around 500 hour. Hence, we can say that best time of the day for departure is nearly around 500 hour.
